```python
import websocket
import json
from kafka import KafkaProducer
from datetime import datetime

# Kafka producer configuration
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: str(v).encode('utf-8')
)

# Variable to store the last price
last_price = None

# Function to send price to Kafka
def send_to_kafka(price, timestamp):
    message = {
        'price': price,
        'timestamp': timestamp
    }
    producer.send('bitcoin-price-topic', value=message)
    producer.flush()

# Function to process incoming WebSocket messages
def on_message(ws, message):
    global last_price  # Declare the variable as global to modify it
    data = json.loads(message)
    price = float(data['p'])  # Get price from the message and convert to float
    current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    # Check if the price has changed
    if last_price is None or price != last_price:
        print(f"Current BTC Price: {price}")
        send_to_kafka(price, current_time)
        last_price = price  # Update the last price

# Function to handle WebSocket errors
def on_error(ws, error):
    print(f"Error: {error}")

# Function to handle WebSocket close
def on_close(ws):
    print("WebSocket closed")

# Function to handle WebSocket open
def on_open(ws):
    print("WebSocket connected")

# Connect to Binance WebSocket
def run_websocket():
    websocket_url = "wss://stream.binance.com:9443/ws/btcusdt@trade"  # BTC/USDT trade data
    ws = websocket.WebSocketApp(
        websocket_url,
        on_message=on_message,
        on_error=on_error,
        on_close=on_close
    )
    ws.on_open = on_open
    ws.run_forever()

if __name__ == "__main__":
    run_websocket()
```
