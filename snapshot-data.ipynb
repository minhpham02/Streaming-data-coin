```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from kafka import KafkaProducer
from datetime import datetime


# Cu hnh Selenium  chy vi ChromeDriver
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

service = Service('E:/Spark_Kafka_Coin/chromedriver-win32/chromedriver.exe')  # Thay th bng ng dn ti ChromeDriver ca bn
driver = webdriver.Chrome(service=service, options=chrome_options)

# M trang web
url = 'https://www.livecoinwatch.com/price/Bitcoin-BTC'
driver.get(url)

producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: str(v).encode('utf-8')
)

# Hm  gi gi BTC n Kafka
def send_to_kafka(price, timestamp):
    message = {
        'price' : price,
        'timestamp' : timestamp
    }
    producer.send('bitcoin-price-topic', value=message)
    producer.flush()


# Hm  ly gi BTC t trang web
def get_btc_price():
    global price_old
    try:
        # Tm phn t cha th span vi class 'price'
        container = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.cion-item'))
        )
        price_tag = container.find_element(By.CSS_SELECTOR, 'span.price')
        price = price_tag.text

        current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

        if price != price_old:
            print(f"Current BTC Price: {price}")
            send_to_kafka(price, current_time)
            price_old = price
    except Exception as e:
        print(f"An error occurred: {e}")

# Cp nht gi BTC lin tc

while True:
    price_old = None
    get_btc_price()

# ng trnh duyt khi hon tt
driver.quit()
```
